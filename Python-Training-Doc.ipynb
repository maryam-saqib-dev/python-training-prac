### Write a Python program to swap two variables without using a temporary variable.

```python
def swapVar(a,b):
  a = a+b
  b = a-b
  a = a-b
  return a,b

a=3
b=4
a,b = swapVar(a,b)
print(f"After Swapping: a = {a}, and b = {b}")
```

### Check if a given number is a palindrome.

```python
def isPalindrome(n):
  nums = []
  while n>0:
    r = n%10
    nums.append(r)
    n = n//10
  if nums == nums[::-1]:
    return True
  else:
    return False

n = 121
print(isPalindrome(n))
```

### Write a program to count vowels, consonants, digits, and special characters in a string.

```python
import string
def countStringCharTypes(s):
  countV,countC,countD,countS = 0,0,0,0
  for c in s:
    if c in string.ascii_letters:
      if c in 'aeiouAEIOU':
        countV=countV+1
      else:
        countC=countC+1
    elif c.isdigit():
      countD=countD+1
    elif c != ' ':
      countS=countS+1
  return countV,countC,countD,countS

s = "Hello, World! 123"
countV,countC,countD,countS = countStringCharTypes(s)
print(f"Vowels: {countV}, Consonants: {countC}, Digits: {countD}, Special Characters: {countS}")
```

### Create a calculator that can perform +, -, *, /, and % operations.

```python
def Calculator(a,b):
  print("Select a number for operation:")
  print("1 for Addition")
  print("2 for Subtraction")
  print("3 for Multiplication")
  print("4 for Division")
  print("5 for Modulus")
  op= input()
  if op == '1':
    return a+b
  elif op == '2':
    return a-b
  elif op == '3':
    return a*b
  elif op == '4':
    return a/b
  elif op == '5':
    return a%b

a = 5
b = 3
print(Calculator(a,b))
```

### Print the first 10 Fibonacci numbers using a loop.

```python
def Fib():
  nums=[]
  for i in range(10):
    if i==0 or i==1:
      nums.append(i)
    else:
      nums.append(nums[i-1]+nums[i-2])
  print(nums)

Fib()
```

### Write a program to find the factorial of a number using recursion.

```python
def Factorial(n):
  if n==0 or n==1:
    return 1
  else:
    return n*Factorial(n-1)

print(Factorial(5))
```

### Check if a number is prime or not.

```python
def isPrime(n):
  for i in range(1,n):
    if i!=1 and i!=n:
      if n%i==0:
        return False
  return True

print(isPrime(13))
print(isPrime(9))
```

### Print a triangle pattern using * of height n.

```python
def nTriangle(n):
  for i in range(n):
    for j in range(n-i-1):
      print(' ',end='')
    for k in range(1,2*(i+1)):
        print('*',end='')
    print()

nTriangle(3)
```

### Write a function that takes a string and returns a dictionary with character counts.

```python
def StrToCharCount(s):
  char_count={}
  for c in s:
    if c not in char_count:
      char_count.update({c:1})
    else:
      char_count[c]+=1
  return char_count

print(StrToCharCount("Hello"))
```

### Create a lambda function that checks if a number is even or odd.

```python
even_odd = lambda n : print('even') if n%2==0 else print('odd')

even_odd(3)
```

### Use map, filter, and reduce to find the sum of squares of even numbers in a list.

```python
from functools import reduce

def sumSqEven(nums):
  return reduce(lambda a,b: a + b, map(lambda n: n**2, filter(lambda n: n%2==0, nums)))

print(sumSqEven([1,2,3,4,5]))
```
