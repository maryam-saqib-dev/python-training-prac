### Write a Python program to swap two variables without using a temporary variable.

```python
def swap_var(a,b):
  a = a+b
  b = a-b
  a = a-b
  return a,b

a=3
b=4
a,b = swap_var(a,b)
print(f"After Swapping: a = {a}, and b = {b}")
```

### Check if a given number is a palindrome.

```python
def is_palindrome(n):
  nums = []
  while n>0:
    r = n%10
    nums.append(r)
    n = n//10
  if nums == nums[::-1]:
    return True
  else:
    return False

n = 121
print(is_palindrome(n))
```

### Write a program to count vowels, consonants, digits, and special characters in a string.

```python
import string
def count_string_char_types(s):
  countV,countC,countD,countS = 0,0,0,0
  for c in s:
    if c in string.ascii_letters:
      if c in 'aeiouAEIOU':
        countV=countV+1
      else:
        countC=countC+1
    elif c.isdigit():
      countD=countD+1
    elif c != ' ':
      countS=countS+1
  return countV,countC,countD,countS

s = "Hello, World! 123"
countV,countC,countD,countS = count_string_char_types(s)
print(f"Vowels: {countV}, Consonants: {countC}, Digits: {countD}, Special Characters: {countS}")
```

### Create a calculator that can perform +, -, *, /, and % operations.

```python
def calculator(a,b):
  print("Select a number for operation:")
  print("1 for Addition")
  print("2 for Subtraction")
  print("3 for Multiplication")
  print("4 for Division")
  print("5 for Modulus")
  op= input()
  if op == '1':
    return a+b
  elif op == '2':
    return a-b
  elif op == '3':
    return a*b
  elif op == '4':
    return a/b
  elif op == '5':
    return a%b

a = 5
b = 3
print(calculator(a,b))
```

### Print the first 10 Fibonacci numbers using a loop.

```python
def fib():
  nums=[]
  for i in range(10):
    if i==0 or i==1:
      nums.append(i)
    else:
      nums.append(nums[i-1]+nums[i-2])
  print(nums)

fib()
```

### Write a program to find the factorial of a number using recursion.

```python
def factorial(n):
  if n==0 or n==1:
    return 1
  else:
    return n*factorial(n-1)

print(factorial(5))
```

### Check if a number is prime or not.

```python
def is_prime(n):
  for i in range(1,n):
    if i!=1 and i!=n:
      if n%i==0:
        return False
  return True

print(is_prime(13))
print(is_prime(9))
```

### Print a triangle pattern using * of height n.

```python
def n_triangle(n):
  for i in range(n):
    for j in range(n-i-1):
      print(' ',end='')
    for k in range(1,2*(i+1)):
        print('*',end='')
    print()

n_triangle(3)
```

### Write a function that takes a string and returns a dictionary with character counts.

```python
def str_to_char_count(s):
  char_count={}
  for c in s:
    if c not in char_count:
      char_count.update({c:1})
    else:
      char_count[c]+=1
  return char_count

print(str_to_char_count("Hello"))
```

### Use map, filter, and reduce to find the sum of squares of even numbers in a list.

```python
from functools import reduce

def sum_sq_even(nums):
  return reduce(lambda a,b: a + b, map(lambda n: n**2, filter(lambda n: n%2==0, nums)))

print(sum_sq_even([1,2,3,4,5]))
```

### Remove duplicates from a list without using set.

```python
def remove_dup(l):
  dup= []
  for c in l:
    if c in dup:
      l.remove(c)
    else:
      dup.append(c)
  return l

print(remove_dup([1,2,3,4,4,5]))
```

### Merge two dictionaries and sort by keys.

```python
def merge_dict(dict1,dict2):
  rdict={}
  for key,val in dict1.items():
    rdict.update({key:val})
  for key, val in dict2.items():
    rdict.update({key:val})
  rdict = dict(sorted(rdict.items()))
  return rdict

print(merge_dict({'a':1,'e':3},{'t':4,'b':3}))
```

### Create a program that counts word frequency in a paragraph.

```python
import string

def word_freq(para):
    words = {}
    s = ''
    for c in para:
        if c != ' ' and c not in string.punctuation:
            s += c
        elif s and s.lower() not in words:
            word = s.lower()
            words.update({word: 1})
            s = ''
        elif s:
          word = s.lower()
          words[word] += 1
          s = ''
    if s.lower() in words:
      word = s.lower()
      words[word] += 1
    else:
       word = s.lower()
       words.update({word: 1})
    return words

print(word_freq("Hello, World! Hello, World"))
```

### Implement a basic student record system using dictionaries.

```python
student_records = {}

def add_student(id, name, age, grade):
    student_records[id] = {'name': name, 'age': age, 'grade': grade}

def delete_student(id):
    if id in student_records:
        del student_records[id]

def view_student(id):
  if id in student_records:
    print(student_records[id])
  else:
    print('Student not found!')

def update_student(id, name=None, age=None, grade=None):
  if id in student_records:
    if name:
      student_records[id]['name'] = name
    if age:
      student_records[id]['age'] = age
    if grade:
      student_records[id]['grade'] = grade
  else:
    print('Student not found!')

add_student(2712,'Maryam',21,'A')
view_student(2712)
delete_student(2712)
view_student(2712)
add_student(2005,'Khadija',20,'B')
view_student(2005)
update_student(2005,'Ali')
view_student(2005)
```
